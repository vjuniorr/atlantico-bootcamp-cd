# Analise exploratoria - IT Salary Survey 2018 - 2020

## Importação das bibliotecas

```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

## Carregando os dados

```python
salary_2018 = pd.read_csv("../../data/raw/IT-Salary-Survey-EU-2018.csv")
salary_2019 = pd.read_csv("../../data/raw/IT-Salary-Survey-EU-2019.csv")
salary_2020 = pd.read_csv("../../data/raw/IT-Salary-Survey-EU-2020.csv")
```

## Perguntas de partida e hipotese

```python
## Criando uma nova coluna de salario, onde o salario é dividido por mil
salary_2018 = salary_2018.assign(salario_mil = salary_2018["Current Salary"] / 1000)

## Setando o estilo do seaborn
sns.set_style("whitegrid")

## Definindo o tamanho da figura e configurando o barplot
plt.figure(figsize=(6, 4))
ax = sns.barplot(data=salary_2018, x="salario_mil", y="Main language at work", color="darkcyan", ci=None)

## Passando as configurações de variavel para dentro da barra
ax.bar_label(
    ax.containers[0],
    fmt='%.1f',
    label_type='center',
    color='white'
)

## Setando as labels do grafico
plt.xlabel("Salario anual (Mil dolares)")
plt.ylabel("Linguagem usada no trabalho")
plt.title("Salario com base na linguagem")

## Removendo as bordas e adicionando uma linha vertical no ponto inicial
ax.set(frame_on=False)
ax.axvline(0, color='k', clip_on=False)
plt.show()
```

```python
## Criando uma nova coluna de salario, onde o salario é dividido por mil
salary_2019 = salary_2019.assign(salario_mil = salary_2019["Yearly brutto salary (without bonus and stocks)"] / 1000)

## Setando o estilo do seaborn
sns.set_style("whitegrid")

## Definindo o tamanho da figura e configurando o barplot
plt.figure(figsize=(6, 6))
ax = sns.barplot(data=salary_2019, x="salario_mil", y="Main language at work", color="darkcyan", ci=None)

## Passando as configurações de variavel para dentro da barra
ax.bar_label(
    ax.containers[0],
    fmt='%.1f',
    label_type='center',
    color='white'
)

## Setando as labels do grafico
plt.xlabel("Salario anual (Mil dolares)")
plt.ylabel("Linguagem usada no trabalho")
plt.title("Salario com base na linguagem")

## Removendo as bordas e adicionando uma linha vertical no ponto inicial
ax.set(frame_on=False)
ax.axvline(0, color='k', clip_on=False)
plt.show()
```

```python
## Criando uma nova coluna de salario, onde o salario é dividido por mil
salary_2020 = salary_2020.assign(salario_mil = salary_2020["Yearly brutto salary (without bonus and stocks) in EUR"] / 1000)

## Setando o estilo do seaborn
sns.set_style("whitegrid")

## Definindo o tamanho da figura e configurando o barplot
plt.figure(figsize=(6, 4))
ax = sns.barplot(data=salary_2020, x="salario_mil", y="Main language at work", color="darkcyan", ci=None)

## Passando as configurações de variavel para dentro da barra
ax.bar_label(
    ax.containers[0],
    fmt='%.1f',
    label_type='center',
    color='white'
)

## Setando as labels do grafico
plt.xlabel("Salario anual (Mil dolares)")
plt.ylabel("Linguagem usada no trabalho")
plt.title("Salario com base na linguagem")

## Removendo as bordas e adicionando uma linha vertical no ponto inicial
ax.set(frame_on=False)
ax.axvline(0, color='k', clip_on=False)
plt.show()
```
